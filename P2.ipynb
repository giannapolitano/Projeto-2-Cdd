```python
def replace_words(lista_principal, wtf, wtf2):
    for lista in lista_principal:
        for palavra in lista:
            for e in wtf:
                for palavra in lista:
                    if e in palavra:
                        palavra2 = palavra.replace(e, '')
                        lista.remove(palavra)
                        lista.append(palavra2)
        for palavra in lista:
            for f in wtf2:
                if f in palavra:
                    lista.remove(palavra)
        for palavra in lista:
            if palavra == 'vc':
                palavra = palavra.replace('vc','voce')
            if palavra == 'qm':
                palavra = palavra.replace('qm','quem')
```

```python
wtf_list = [':','"', "'", '/', '...', ',', ';', '#', '$', '!', '?', '.', '‚Ä¶', '\n']
wtf_list2 = ['http', '@']
```

```python
replace_words(tweets, wtf_list, wtf_list2)
replace_words(lista_tweets_relevantes, wtf_list, wtf_list2)
replace_words(lista_tweets_irrelevantes, wtf_list, wtf_list2)
```

```python
palavras = {}
for tweet in tweets:
    for palavra in tweet:
        if palavra not in palavras:
            palavras[palavra] = 1
        else:
            palavras[palavra] += 1
            
palavras_relevantes = {}
for tweet in lista_tweets_relevantes:
    for palavra in tweet:
        if palavra not in palavras_relevantes:
            palavras_relevantes[palavra] = 1
        else:
            palavras_relevantes[palavra] += 1
            
palavras_irrelevantes = {}
for tweet in lista_tweets_irrelevantes:
    for palavra in tweet:
        if palavra not in palavras_irrelevantes:
            palavras_irrelevantes[palavra] = 1
        else:
            palavras_irrelevantes[palavra] += 1
```
